```python
def monthly_payment(principal, interest_as_decimal, years):
    remaining_bal=principal
    i=0
    while remaining_bal >=0:
        number_payments = years*12
        monthslist = list(range(number_payments))
        rate = interest_as_decimal/12
        numerator=principal*(rate)*((1+rate)**number_payments)
        denominator = ((1+rate)**number_payments)-1

        monthly_pay = numerator/denominator
        remaining_bal = remaining_bal-monthly_pay
        interest = remaining_bal*rate
        payment=monthly_pay+interest
        if i < len(monthslist):
            print("|",monthslist[i],"|",round(monthly_pay,2),"|",round(interest,2),"|",round(payment,2),"|",round(remaining_bal,2),"|")
            i=+ i+1
    return monthly_pay
```

Plots a normal distribution around given statistics

```python

def norm (mean=10,stdev=1,npoints=[1,10,100,1000,10000,1000000],nbins=20):
    import numpy as np
    import matplotlib.pyplot as plt
    import pandas as pd
    plt.xlabel('Value')  
    plt.ylabel('Frequency') 
    plt.tight_layout()
    
    longest = 1
    if type(mean)==list:
        if longest<len(mean):
            longest = len(mean)
            for i in range(longest):
                plt.subplot(int(round(longest/2,0)),int(round(longest/2,0)),i+1)
                x = np.random.normal(mean[i], stdev, npoints)
                plt.hist(x, bins=nbins)
                plt.title(str(mean[i])+' Mean')
                
    elif type(stdev)==list:
        if longest<len(stdev):
            longest = len(stdev)
            for i in range(longest):
                plt.subplot(int(round(longest/2,0)),int(round(longest/2,0)),i+1)
                x = np.random.normal(mean, stdev[i], npoints)
                plt.hist(x, bins=nbins)
                plt.title(str(stdev[i])+' Standard Deviation') 

    elif type(npoints)==list:
        if longest<len(npoints):
            longest = len(npoints)
            for i in range(longest):
                plt.subplot(int(round(longest/2,0)),int(round(longest/2,0)),i+1)
                x = np.random.normal(mean, stdev, npoints[i])
                plt.hist(x, bins=nbins)
                plt.title(str(npoints[i])+' Points')

    elif type(nbins)==list:
        if longest<len(nbins):
            longest = len(nbins)
            for i in range(longest):
                plt.subplot(int(round(longest/2,0)),int(round(longest/2,0)),i+1)
                x = np.random.normal(mean, stdev, npoints)
                plt.hist(x, bins=nbins[1])
                plt.title(str(nbins[i])+' Bins')
    else:
        x = np.random.normal(mean, stdev, npoints)
        plt.hist(x, bins=nbins)
            
    plt.show()
    plotting=':)'
    return plotting
```

Lin Cubic and Sqrt Equations for correspondance

```python
def linear_eq(x, a, b):
    """
    Calculate the value of a linear equation y = a + bR.

    Args:
        x (float): The value of the independent variable.
        a (float): The constant term in the linear equation.
        b (float): The coefficient of x in the linear equation.

    Returns:
        float: The value of the linear equation y = ax + b, evaluated at the given x.
    """
    y = a + b * x
    return y


def cubic_eq(x, a, b, c, d):
    """
    Calculate the value of a cubic equation y = a + b*x + c * x^2 + d * x^3.

    Args:
        x (float): The value of the independent variable.
        a (float): Coefficient of the bias term.
        b (float): Ceofficient of the linear term.
        c (float): Coefficient of the quadratic term.
        d (float): Coefficient of the cubic term.

    Returns:
        float: The value of the cubic equation, evaluated at the given x.

    """
    y = 0
    coeff = [a, b, c, d]
    for i in range(len(coeff)):
        # Iterate through the coefficients and calculate the corresponding term
        y += coeff[i] * x ** i

    return y

def sqrt_eq(x, a, b):
    """
    Calculate the value of a linear equation y = a + bR.

    Args:
        x (float): The value of the independent variable.
        a (float): The constant term in the linear equation.
        b (float): The coefficient of x in the linear equation.

    Returns:
        float: The value of the linear equation y = ax + b, evaluated at the given x.
    """
    y = a + b * np.sqrt(x)
    return y
```

Assorted Statistical Analysis

```python
def calculate_r_squared(y_true, y_pred):
    """
    Calculate R-squared score.

    Parameters:
    - y_true: array-like, true target values.
    - y_pred: array-like, predicted target values by the regression model.

    Returns:
    - R-squared value.
    """
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    r_squared = 1 - (ss_res / ss_tot)
    return r_squared

def calculate_mse(y_true, y_pred):
    """
    Calculate Mean Squared Error (MSE).

    Parameters:
    - y_true: array-like, true target values.
    - y_pred: array-like, predicted target values by the regression model.

    Returns:
    - Mean Squared Error value.
    """
    mse = np.mean((y_true - y_pred) ** 2)
    return mse

def calculate_mae(y_true, y_pred):
    """
    Calculate Mean Absolute Error (MAE).

    Parameters:
    - y_true: array-like, true target values.
    - y_pred: array-like, predicted target values by the regression model.

    Returns:
    - Mean Absolute Error value.
    """
    mae = np.mean(np.abs(y_true - y_pred))
    return mae

def calculate_rmse(y_true, y_pred):
    """
    Calculate Root Mean Squared Error (RMSE).

    Parameters:
    - y_true: array-like, true target values.
    - y_pred: array-like, predicted target values by the regression model.

    Returns:
    - Root Mean Squared Error value.
    """
    rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))
    return rmse
def model_selection(ydata, y_linear_fit,y_cubic_fit,y_sqrt_fit):
    # Calculate RMSE for each fit
    rmse_linear = calculate_rmse(ydata, y_linear_fit)
    rmse_cubic = calculate_rmse(ydata, y_cubic_fit)
    rmse_sqrt = calculate_rmse(ydata, y_sqrt_fit)

    # Calculate MAE for each fit
    mae_linear = calculate_mae(ydata, y_linear_fit)
    mae_cubic = calculate_mae(ydata, y_cubic_fit)
    mae_sqrt = calculate_mae(ydata, y_sqrt_fit)

    # Calculate R-squared for each fit
    r_squared_linear = calculate_r_squared(ydata, y_linear_fit)
    r_squared_cubic = calculate_r_squared(ydata, y_cubic_fit)
    r_squared_sqrt = calculate_r_squared(ydata, y_sqrt_fit)
    
    # Prepare the data for the heatmap
    metrics_data = {
        'RÂ²': [r_squared_linear, r_squared_cubic, r_squared_sqrt],
        'MAE': [mae_linear, mae_cubic, mae_sqrt],
        'RMSE': [rmse_linear, rmse_cubic, rmse_sqrt]
    }
    metrics_df = pd.DataFrame(metrics_data, index=['Linear', 'Cubic', 'Sqrt'])

    # Plotting the heatmap
    plt.figure(figsize=(10, 3))
    sns.heatmap(metrics_df, annot=True, cmap="YlGnBu", fmt=".2f")
    plt.title('Performance Metrics for Different Fits')
    plt.show()
```
